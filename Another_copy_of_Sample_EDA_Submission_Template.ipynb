## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Fill missing values in categorical columns with Mode

df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])
df['Dosage'] = df['Dosage'].fillna(df['Dosage'].mode()[0])

# Fill missing values in numerical column with Median
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(df['Line Item Insurance (USD)'].median())
````

````python
df.isnull().sum()
````

### What all manipulations have you done and insights you found?

Checked the dataset structure and data types.

Verified duplicate records and found no duplicate rows.

Identified missing values in Shipment Mode, Dosage, and Line Item Insurance (USD).

Replaced missing categorical values using the mode.

Replaced missing numerical values using the median.

Verified that the dataset contains no missing values after preprocessing.

Prepared the cleaned dataset for visualization and analysis.
